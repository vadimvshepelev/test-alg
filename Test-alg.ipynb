```python
import numpy as np
from scipy.optimize import minimize

math_exp, disp = 10, 10
np.random.normal(math_exp, disp, 1)


def calc_data(K: float):
    n_max = 100
    data = np.zeros(n_max)
    n = list(range(n_max))
    # K = -1
    data[0] = 0. + .3*.5
    for i in range(1, n_max):
        x_new_1 = .2*data[i-1]
        x_new_2 = .3*np.random.normal(math_exp, disp, 1) 
        x_new_3 = .4*K*data[i-1]
        data[i] = x_new_1 + x_new_2 + x_new_3
    
    print(data)
    
    ctrl_value = sum([elem*elem for elem in data])
    print(ctrl_value)
    return ctrl_value

b = (-1000, 1000)

# sol = minimize(calc_data, x0=0, method='COBYLA', bounds=(b,))
sol = minimize(calc_data, x0=0, method='BFGS', bounds=(b,))
sol



```

```python
import matplotlib.pyplot as plt
%matplotlib inline

res, data = calc_data(-1) 

plt.title("Последовательность X_n") # заголовок
plt.xlabel("n") # ось абсцисс
plt.ylabel("x") # ось ординат
plt.grid()      # включение отображение сетки
plt.plot(n, data)  # построение графика

```
